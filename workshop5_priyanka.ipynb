To-Do-4
Cost Function (MSE)

```python
def cost_function(X, Y, W):
    m = len(Y)
    Y_pred = X.dot(W)
    cost = np.sum((Y_pred - Y) ** 2) / (2 * m)
    return cost
```

```python
X_test_case = np.array([[1, 2], [3, 4], [5, 6]])
Y_test_case = np.array([3, 7, 11])
W_test_case = np.array([1, 1])

cost = cost_function(X_test_case, Y_test_case, W_test_case)

if cost == 0:
    print("Proceed Further")
else:
    print("Something went wrong")

print("Cost:", cost)
```

To-Do-6: Gradient Descent Implementation

```python
def gradient_descent(X, Y, W, alpha, iterations):
    m = len(Y)
    cost_history = []

    for _ in range(iterations):
        Y_pred = X.dot(W)
        loss = Y_pred - Y
        dw = (X.T.dot(loss)) / m
        W = W - alpha * dw
        cost_history.append(cost_function(X, Y, W))

    return W, cost_history
```

To-Do-7: Test Gradient Descent

```python
np.random.seed(0)
X_rand = np.random.rand(100, 3)
Y_rand = np.random.rand(100)
W_rand = np.random.rand(3)

alpha = 0.01
iterations = 1000

final_W, cost_hist = gradient_descent(X_rand, Y_rand, W_rand, alpha, iterations)

print("Final Parameters:", final_W)
print("Final Cost:", cost_hist[-1])
```

To-Do-8: RMSE Function

```python
def rmse(Y, Y_pred):
    return np.sqrt(np.mean((Y - Y_pred) ** 2))
```

To-Do-9: R² Function (From Scratch)

```python
def r2(Y, Y_pred):
    mean_y = np.mean(Y)
    ss_tot = np.sum((Y - mean_y) ** 2)
    ss_res = np.sum((Y - Y_pred) ** 2)
    return 1 - (ss_res / ss_tot)
```

To-Do-10: Main Function (Complete Workflow)

```python
def main():
    data = pd.read_csv("student.csv")

    X = data[['Math', 'Reading']].values
    Y = data['Writing'].values

    # Train-test split
    np.random.seed(42)
    indices = np.random.permutation(len(X))
    train_size = int(0.8 * len(X))
    train_idx = indices[:train_size]
    test_idx = indices[train_size:]

    X_train = X[train_idx]
    X_test = X[test_idx]
    Y_train = Y[train_idx]
    Y_test = Y[test_idx]

    # Initialize
    W = np.zeros(X_train.shape[1])
    alpha = 0.00001
    iterations = 1000

    # Train
    W_opt, cost_hist = gradient_descent(X_train, Y_train, W, alpha, iterations)

    # Predict
    Y_pred = X_test.dot(W_opt)

    # Evaluate
    print("Final Weights:", W_opt)
    print("Cost (first 10):", cost_hist[:10])
    print("RMSE:", rmse(Y_test, Y_pred))
    print("R2 Score:", r2(Y_test, Y_pred))
```

```python
if __name__ == "__main__":
    main()
```
